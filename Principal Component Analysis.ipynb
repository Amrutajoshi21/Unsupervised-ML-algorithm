```python
ppt = ppt.drop({'Sale_Price'}, axis=1)
```

before applying PCA we need to make data unitless : standardization

```python
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_ppt = scaler.fit_transform(ppt)
scaler_ppt
```

```python
scaler_ppt.shape
```

```python
# data has been converted unitless
```

```python
pca = PCA()
```

```python
pca_ppt = pca.fit_transform(scaler_ppt)
```

```python
pca.explained_variance_ratio_
```

```python
# these are principle component assigned to every column in ascending order
list(pca.explained_variance_ratio_)
```

```python
# 1 is divided into 74 value according to importance of columns
# default error is 5%, ignore them pcas
# we are extracting only those no of components that covering around 95% of variance
```

```python
list = list(pca.explained_variance_ratio_)
```

```python
np.sum(list[0:55])
```

```python
#pca can be used for dimensionality reduction not for feature selection
```

```python
df = pd.DataFrame(pca_ppt)
df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(df.iloc[:,0:55],ppt1.Sale_Price)
```

```python
lr.score(df.iloc[:,0:55],ppt1.Sale_Price)
```
